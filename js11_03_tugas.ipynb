```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/wbc.csv')

# slice kolom terakhir --> Unnamed column
df = df.iloc[:,:-1]

df.head()
```

#### Seleksi Fitur

```python
X = df.iloc[:, 2:]
y = df['diagnosis']

X.head()
```

#### Split Data Training dan Testing

```python
from sklearn.model_selection import RepeatedKFold

# inisiasi repated k-fold
cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)
```

#### 1. Random Search

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier

# Inisiasi Model
dt = DecisionTreeClassifier()

# Hyperparameter Space
rs_space = {    
    'max_depth': list(range(5,11)),
    'criterion':['gini', 'entropy']
}

## Inisiasi Randomized Search
rand =  RandomizedSearchCV(dt, rs_space, n_iter=500, scoring='accuracy', n_jobs=-1, cv=cv)
model_random = rand.fit(X,y)

## Print Hasil Best Score & Best Params
print(f'Hasil evaluasi: {model_random.best_score_}')
print(f'Konfigurasi hyperparameter: {model_random.best_params_}')
```

#### 2. Tunning SVM Algorithm

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# inisiasi model
model = SVC()

# Hyperparameter
param_grid = {
    'C': [0.1], 
    'degree': [3],
    'kernel': ['rbf']
} 

# Inisiasi GridSearchCV dengan Alogritma SVM
grid = GridSearchCV(SVC(), param_grid, cv=cv)

%timeit grid.fit(X, y)
# Evaluasi dengan score
score = grid.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {grid.best_params_}')
```
